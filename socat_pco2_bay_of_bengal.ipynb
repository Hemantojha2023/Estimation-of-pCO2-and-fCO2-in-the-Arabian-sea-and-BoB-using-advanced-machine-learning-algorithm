__Let's Impliment XGBoost on Bay of Bengal for pCO2__

```python
import pandas as pd
df1=pd.read_csv('pCO2_Bay_of_Bengal.csv')
```

```python
df1
```

```python
df1=df1.drop('Unnamed: 0', axis=1)
```

```python
df1.isna().sum()
```

```python
df1['year'].value_counts()
```

```python
train_df = df1[df1['year'] != 2016].reset_index()
test_df = df1[df1['year'] == 2016].reset_index()
```

```python
train_df['year'].value_counts()
```

```python
train_df.drop(['index', 'year', 'Lat', 'Lon', 'month', 'day', 'hour', 'minute', 'second'], axis=1, inplace=True)
```

```python
train_df
```

```python
train_df.isna().sum()
```

```python
test_df.drop(['index', 'year', 'Lat', 'Lon', 'month', 'day', 'hour', 'minute', 'second'], axis=1, inplace=True)
```

```python
test_df
```

```python
test_df.isna().sum()
```

```python
x=train_df.drop('pCO2water_SST_wet', axis=1)
y=train_df['pCO2water_SST_wet']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=123)
```

```python
x_train.shape
```

```python
test_df.shape
```

```python
median_pCO2=y_train.median(skipna=True)
y_train=y_train.fillna(median_pCO2)
y_test=y_test.fillna(median_pCO2)
```

```python
import xgboost as xgb

# Create DMatrix for XGBoost
dtrain = xgb.DMatrix(x_train, label=y_train)
dval = xgb.DMatrix(x_test, label=y_test)
```

```python
# Define parameters
params = {
    "objective": "reg:squarederror",
    "eval_metric": "rmse",
    "learning_rate": 0.1,
    "max_depth": 4,
    #'alpha': 1.5,         # L1 regularization
    #'lambda' : 3.0
}
```

```python
# Train with early stopping
evals = [(dtrain, "train"), (dval, "validation")]
model = xgb.train(
    params, 
    dtrain, 
    num_boost_round=1000,  # Max rounds
    evals=evals,
    early_stopping_rounds=5,  # Stop if no improvement in last 5 rounds
    verbose_eval=True
)
```

```python
# Get the best iteration
best_iteration = model.best_iteration
print(f"Best iteration: {best_iteration}")
```

```python
# Separate target variable from test_df for predictions
test_features=test_df.drop('pCO2water_SST_wet', axis=1)
test_target=test_df['pCO2water_SST_wet']
```

```python
# Fill missing values with median
median_pCO2 = test_target.median(skipna=True)
test_target.fillna(median_pCO2, inplace=True)
```

```python
dval1 = xgb.DMatrix(test_features, label=test_target)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
#y_train_pred=model.predict(dtrain, iteration_range=(0, best_iteration))
# Make predictions
y_pred = model.predict(dval1, iteration_range=(0, best_iteration))

#training_rmse=np.sqrt(mean_squared_error(y_train, y_train_pred))
testing_rmse=np.sqrt(mean_squared_error(test_target, y_pred))

print('XGBoost Regression :')
#print('Training RMSE :',training_rmse)
print('Testing RMSE :',testing_rmse)
```

For DMatrix api we can not use cross-validation because for cross-validation model should be trained using .fit

```python
import xgboost as xgb
import matplotlib.pyplot as plt
plt.figure(figsize=(7, 6))
xgb.plot_importance(model)
plt.savefig('feature_pco2_bob.png')
plt.show()
```

Plot actual vs predicted data points with best fit line 

- If the model is perfect, points will lie on a 45-degree diagonal line.

```python
import seaborn as sns
import matplotlib.pyplot as plt
#plt.figure(figsize=(5, 3))
sns.regplot(x=test_target, y=y_pred, scatter_kws={"color": "blue", "alpha": 0.6}, line_kws={"color": "red"}, ci=None)
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.title("Least Squares Approximation (Actual vs. Predicted)")
plt.savefig('pco2_bob.png')
plt.show()
```

We can see that the most point lies on the regressoin line so this line is the best fit line and model best fit model

__Let's impliment XGBRegressor on the data which has 15% nan value__

```python
x=new_df.drop('pCO2water_SST_wet', axis=1)
y=new_df['pCO2water_SST_wet']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test1 = train_test_split(x, y, test_size=0.2, random_state=42)
```

XGBoost handles missing values in features (X) automatically. But, the target variable (y) cannot contain NaN values. XGBoost does not accept NaN values in y_train

According to standard machine learning practices, filling missing values in the test set with statistics (like median) calculated from the training set does not constitute Data leakage. This method ensures that no information from the test set influences the training process, thus maintaining the integrity of the evaluation.

```python
median_pCO2=y_train.median(skipna=True)
y_train=y_train.fillna(median_pCO2)
y_test1=y_test1.fillna(median_pCO2)
```

Use XGBoost Native API offers more control over model training and customization.

```python
import xgboost as xgb

# Create DMatrix for XGBoost
dtrain = xgb.DMatrix(x_train, label=y_train)
dval = xgb.DMatrix(x_test, label=y_test1)
```

```python
print(dtrain.num_row(), dtrain.num_col()) 
```

```python
# Define parameters
params = {
    "objective": "reg:squarederror",
    "eval_metric": "rmse",
    "learning_rate": 0.1,
    "max_depth": 5
}
```

```python
# Callbacks for early stopping and logging
# callbacks = [
#     xgb.callback.EarlyStopping(rounds=10, metric_name="rmse", data_name="validation"),  # Stops after 10 rounds of no improvement
#     xgb.callback.LearningRateScheduler(lambda epoch: 0.1 * (0.99 ** epoch)),  # Custom learning rate decay
#     xgb.callback.LogEvaluation(period=5)  # Log every 5 rounds
```

```python
# Train with early stopping
evals = [(dtrain, "train"), (dval, "validation")]
model = xgb.train(
    params, 
    dtrain, 
    num_boost_round=1000,  # Max rounds
    evals=evals,
    early_stopping_rounds=10,  # Stop if no improvement in last 10 rounds
    verbose_eval=True
)
```

```python
# Get the best iteration
best_iteration = model.best_iteration
print(f"Best iteration: {best_iteration}")
```

```python
# Make predictions
y_pred1 = model.predict(dval, iteration_range=(0, best_iteration))
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
y_train_pred=model.predict(dtrain, iteration_range=(0, best_iteration))


training_rmse=np.sqrt(mean_squared_error(y_train, y_train_pred))
testing_rmse=np.sqrt(mean_squared_error(y_test1, y_pred1))

print('XGBoost Regression :')
print('Training RMSE :',training_rmse)
print('Testing RMSE :',testing_rmse)
```

model is generalize well on training set but not on testing set so model overfitted
