```python
import geopandas as gpd
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from shapely.geometry import Point

# Convert DataFrame to GeoDataFrame
features['geometry'] = features.apply(lambda row: Point(row['lon'], row['lat']), axis=1)
gdf = gpd.GeoDataFrame(features, geometry='geometry')

# Define boundaries for Arabian Sea and Bay of Bengal
gdf_arabian_sea = gdf.cx[51:75, -1:26]
gdf_bay_of_bengal = gdf.cx[78:96, 5:25]

# Find common vmin and vmax for Salinity and SST to ensure consistent color scales
min_salinity = min(gdf_arabian_sea['Salinity'].min(), gdf_bay_of_bengal['Salinity'].min())
max_salinity = max(gdf_arabian_sea['Salinity'].max(), gdf_bay_of_bengal['Salinity'].max())

min_sst = min(gdf_arabian_sea['SST'].min(), gdf_bay_of_bengal['SST'].min())
max_sst = max(gdf_arabian_sea['SST'].max(), gdf_bay_of_bengal['SST'].max())

min_pressure = min(gdf_arabian_sea['Atm_Pressure'].min(), gdf_bay_of_bengal['Atm_Pressure'].min())
max_pressure = max(gdf_arabian_sea['Atm_Pressure'].max(), gdf_bay_of_bengal['Atm_Pressure'].max())

min_pco2 = min(gdf_arabian_sea['pCO2'].min(), gdf_bay_of_bengal['pCO2'].min())
max_pco2 = max(gdf_arabian_sea['pCO2'].max(), gdf_bay_of_bengal['pCO2'].max())

min_fco2 = min(gdf_arabian_sea['fCO2'].min(), gdf_bay_of_bengal['fCO2'].min())
max_fco2 = max(gdf_arabian_sea['fCO2'].max(), gdf_bay_of_bengal['fCO2'].max())

# Create a plot with Cartopy projections
fig, axes = plt.subplots(5, 2, figsize=(15, 21), subplot_kw={'projection': ccrs.PlateCarree()})

# Define common plotting options
extent = [50, 96, -2, 26]  # Extent covering both regions

# Plot Salinity for Arabian Sea (Top Left)
axes[0, 0].set_extent(extent)
axes[0, 0].add_feature(cfeature.COASTLINE)
sc1 = axes[0, 0].scatter(gdf_arabian_sea['lon'], gdf_arabian_sea['lat'], c=gdf_arabian_sea['Salinity'],
                         cmap='viridis', vmin=min_salinity,vmax=max_salinity)
axes[0, 0].set_title("Arabian Sea - Salinity")
plt.colorbar(sc1, ax=axes[0, 0], orientation='horizontal',label="Salinity (ppt)", shrink=0.54)

# Plot Salinity for Bay of Bengal (Top Right)
axes[0, 1].set_extent(extent)
axes[0, 1].add_feature(cfeature.COASTLINE)
sc2 = axes[0, 1].scatter(gdf_bay_of_bengal['lon'], gdf_bay_of_bengal['lat'], c=gdf_bay_of_bengal['Salinity'],
                         cmap='viridis', vmin=min_salinity,vmax=max_salinity)
axes[0, 1].set_title("Bay of Bengal - Salinity")
plt.colorbar(sc2, ax=axes[0, 1], orientation='horizontal',label="Salinity (ppt)", shrink=0.54)

# Plot SST for Arabian Sea (middle Left)
axes[1, 0].set_extent(extent)
axes[1, 0].add_feature(cfeature.COASTLINE)
sc3 = axes[1, 0].scatter(gdf_arabian_sea['lon'], gdf_arabian_sea['lat'], c=gdf_arabian_sea['SST'],
                         cmap='coolwarm',vmin=min_sst,vmax=max_sst)
axes[1, 0].set_title("Arabian Sea - SST")
plt.colorbar(sc3, ax=axes[1, 0], orientation='horizontal',label="SST (°C)", shrink=0.54)

# Plot SST for Bay of Bengal (middle Right)
axes[1, 1].set_extent(extent)
axes[1, 1].add_feature(cfeature.COASTLINE)
sc4 = axes[1, 1].scatter(gdf_bay_of_bengal['lon'], gdf_bay_of_bengal['lat'], c=gdf_bay_of_bengal['SST'],
                         cmap='coolwarm',vmin=min_sst,vmax=max_sst)
axes[1, 1].set_title("Bay of Bengal - SST")
plt.colorbar(sc4, ax=axes[1, 1], orientation='horizontal',label="SST (°C)", shrink=0.54)

# Plot Atmospheric Pressure for Arabian Sea (Left)
axes[2, 0].set_extent(extent)
axes[2, 0].add_feature(cfeature.COASTLINE)
sc5 = axes[2, 0].scatter(gdf_arabian_sea['lon'], gdf_arabian_sea['lat'], c=gdf_arabian_sea['Atm_Pressure'],
                         cmap='coolwarm',vmin=min_pressure,vmax=max_pressure)
axes[2, 0].set_title("Arabian Sea-Atmospheric Pressure")
plt.colorbar(sc5, ax=axes[2, 0], orientation='horizontal',label="Atm Pressure (hPa)", shrink=0.54)

# Plot Atmospheric Pressure for Bay of Bengal (Right)
axes[2, 1].set_extent(extent)
axes[2, 1].add_feature(cfeature.COASTLINE)
sc6 = axes[2, 1].scatter(gdf_bay_of_bengal['lon'], gdf_bay_of_bengal['lat'], c=gdf_bay_of_bengal['Atm_Pressure'],
                         cmap='coolwarm',vmin=min_pressure,vmax=max_pressure)
axes[2, 1].set_title("Bay of Bengal-Atmospheric Pressure")
plt.colorbar(sc6, ax=axes[2, 1], orientation='horizontal',label="Atm Pressure (hpa)", shrink=0.54)

# Plot pCO2 for Arabian Sea (bottom Left)
axes[3, 0].set_extent(extent)
axes[3, 0].add_feature(cfeature.COASTLINE)
sc7 = axes[3, 0].scatter(gdf_arabian_sea['lon'], gdf_arabian_sea['lat'], c=gdf_arabian_sea['pCO2'],
                         cmap='viridis', vmin=min_pco2,vmax=max_pco2)
axes[3, 0].set_title("Arabian Sea - pCO2")
plt.colorbar(sc7, ax=axes[3, 0], orientation='horizontal',label="pCO2 (uatm)", shrink=0.54)

# Plot pCO2 for Bay of Bengal (bottom Right)
axes[3, 1].set_extent(extent)
axes[3, 1].add_feature(cfeature.COASTLINE)
sc8 = axes[3, 1].scatter(gdf_bay_of_bengal['lon'], gdf_bay_of_bengal['lat'], c=gdf_bay_of_bengal['pCO2'],
                         cmap='viridis', vmin=min_pco2,vmax=max_pco2)
axes[3, 1].set_title("Bay of Bengal - pCO2")
plt.colorbar(sc8, ax=axes[3, 1], orientation='horizontal',label="pCO2 (uatm)", shrink=0.54)

# Plot fCO2 for Arabian Sea (bottom Left)
axes[4, 0].set_extent(extent)
axes[4, 0].add_feature(cfeature.COASTLINE)
sc9 = axes[4, 0].scatter(gdf_arabian_sea['lon'], gdf_arabian_sea['lat'], c=gdf_arabian_sea['fCO2'],
                         cmap='viridis', vmin=min_fco2,vmax=max_fco2)
axes[4, 0].set_title("Arabian Sea - fCO2")
plt.colorbar(sc9, ax=axes[4, 0], orientation='horizontal',label="fCO2 (uatm)", shrink=0.54)

# Plot fCO2 for Bay of Bengal (bottom Right)
axes[4, 1].set_extent(extent)
axes[4, 1].add_feature(cfeature.COASTLINE)
sc10 = axes[4, 1].scatter(gdf_bay_of_bengal['lon'], gdf_bay_of_bengal['lat'], c=gdf_bay_of_bengal['fCO2'],
                         cmap='viridis', vmin=min_fco2,vmax=max_fco2)
axes[4, 1].set_title("Bay of Bengal - fCO2")
plt.colorbar(sc10, ax=axes[4, 1], orientation='horizontal',label="fCO2 (uatm)", shrink=0.54)

# Add additional map features
for ax in axes.flat:
    ax.add_feature(cfeature.BORDERS, linestyle=':')
    ax.gridlines(draw_labels=True, linewidth=0.5, color='gray', linestyle='--')

# Layout adjustments
plt.tight_layout()
plt.savefig('output2.png', dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from matplotlib.patches import Rectangle

# Define overall map extent
extent = [50, 98, 0, 25]  # [lon_min, lon_max, lat_min, lat_max]

# Define rectangle bounds for Arabian Sea and Bay of Bengal
arabian_sea_bounds = [50, 0, 25, 22]  # [lon_min, lat_min, width, height]
bay_of_bengal_bounds = [80, 5, 18, 17]

# Create map
fig, ax = plt.subplots(figsize=(8, 6), subplot_kw={'projection': ccrs.PlateCarree()})
ax.set_extent(extent, crs=ccrs.PlateCarree())

# Add features
ax.add_feature(cfeature.LAND, facecolor='lightgray')
ax.add_feature(cfeature.OCEAN, facecolor='lightblue')
ax.add_feature(cfeature.COASTLINE, linewidth=1)
ax.add_feature(cfeature.BORDERS, linestyle=':', edgecolor='black')
ax.add_feature(cfeature.RIVERS, edgecolor='blue', linewidth=0.5)

# Add rectangular box for Arabian Sea
arabian_box = Rectangle(
    (arabian_sea_bounds[0], arabian_sea_bounds[1]),  # bottom left corner
    arabian_sea_bounds[2], arabian_sea_bounds[3],    # width and height
    linewidth=2, edgecolor='red', facecolor='none', label='Arabian Sea'
)
ax.add_patch(arabian_box)

# Add rectangular box for Bay of Bengal
bengal_box = Rectangle(
    (bay_of_bengal_bounds[0], bay_of_bengal_bounds[1]),
    bay_of_bengal_bounds[2], bay_of_bengal_bounds[3],
    linewidth=2, edgecolor='green', facecolor='none', label='Bay of Bengal'
)
ax.add_patch(bengal_box)

# Gridlines
gl = ax.gridlines(draw_labels=True, linewidth=0.5, color='gray', alpha=0.5, linestyle='--')
gl.top_labels = False
gl.right_labels = False

# Add legend and title
ax.legend(loc='lower left')
ax.set_title("Arabian Sea and Bay of Bengal with Defined Regions", fontsize=16, pad=20)

plt.tight_layout()
plt.savefig('study_region.jpg')
plt.show()
```
